**Q: What is the correct values for the TFC row?**

To calculate Total Final Consumption, we have to add consumption by Industry, Transport and Other sectors:

```python
tfc = df.loc[['Industry', 'Transport', 'Other sectors']].sum()
list(tfc.values)
```

**Q: How much electricity must Coronie produce in 2015? (i.e., what is the value to fill in the row ‘Electricity Plants’, of the column ‘Electricity’)**

You know that no electricity is imported or exported:

```python
df.loc['Imports','Electricity'] = 0
df.loc['Exports','Electricity'] = 0
```

Therefore the TPES of electricity is zero:

```python
df.loc['TPES','Electricity'] = df.loc['Production', 'Electricity'] + df.loc['Imports', 'Electricity'] + df.loc['Exports', 'Electricity'] + df.loc['Stock changes', 'Electricity'] 
```

And we can cacluate what we need to generate from electricity plants:

```python
df.loc['Electricity Plants','Electricity'] = df.loc['TFC','Electricity'] - df.loc['Other energy conversion','Electricity']
```

Add in sum:

```python
df.loc['Electricity Plants','Total'] = df.loc['Electricity Plants'].sum()
```

```python
df
```

**Q: In Coronie, 50% of the electricity is generated in coal fired power plants with an efficiency of 40%. The rest of the electricity is generated in gas fired power plants with an efficiency of 50%.
How much coal and gas is used for the electricity production? Which answer indicates the values to be filled in the "electricity plants" row?**

```python
coalRequired = -( (0.5 * 800) / 0.4 )
gasRequired = -( (0.5 * 800) / 0.5 )
```

```python
df.loc['Electricity Plants','Coal'] = coalRequired
df.loc['Electricity Plants','Natural Gas'] = gasRequired
df
```

**Q: What are the correct values for the TPES row?**

TPES is TFC minus any conversions.

```python
df.loc['Electricity Plants','Natural Gas'] = gasRequired
```

```python
for t in ['Coal','Natural Gas','Electricity']:
    df.loc['TPES', t] = df.loc['TFC', t] -  df.loc['Electricity Plants',t] -  df.loc['Other energy conversion',t]
df.loc['TPES', 'Total'] = df.loc['TPES'].sum()
df
```

Now we can fill in the missing gaps:

```python
df.loc['Imports','Coal'] = df.loc['TPES','Coal']
df.loc['Production','Natural Gas'] =  df.loc['TPES','Natural Gas']
df.loc['Production','Total'] =  df.loc['Production'].sum()
df.loc['Imports','Total'] =  df.loc['Imports'].sum()
df
```

**Q: What is the current energy use in kWh/day per person in the UK for Transport, Industry and Building?**

For this question we will use the **TFC** sector data from [this IEA report](https://www.iea.org/data-and-statistics/data-tools/energy-statistics-data-browser?country=UK&fuel=Energy%20consumption&indicator=TFCShareBySector) which provides the following figures:
* **Residential** = 1585004TJ/year
* **Transport** = 1516124TJ/year
* **Industry** = 900678TJ/year
* **Commercial and Public Services** = 669543TJ/year

In order to calculate kWh/person/day we need to covert from TJ/year to kWh/day and factor in UK population of ~68 million.  The Python GPT from ChatGPT came up with the following function:

```python
def tj_to_kwh_per_day_per_person(tj, population=68_000_000):
    """
    Convert energy from Terajoules (TJ) to Kilowatt-hours (KWh) per day per person.
    
    Parameters:
    tj (float): Energy in Terajoules.
    population (int): Population of the country. Default is 68 million.
    
    Returns:
    float: Energy in Kilowatt-hours per day per person.
    """
    # Convert TJ to KWh
    kwh = tj * 10**12 * 2.77778e-7
    
    # Convert KWh to KWh per day
    kwh_per_day = kwh / 365

    # Convert to KWh per day per person
    kwh_per_day_per_person = kwh_per_day / population
    
    return round(kwh_per_day_per_person,2)
```

Plugging in our values:

```python
vs = [{'Sector':'Residential', 'tj':1585004}, {'Sector':'Transport', 'tj':1516124}, {'Sector':'Industry','tj':900678}, {'Sector':'Commercial & Services','tj':669543} ]
sectors = [v.get('Sector') for v in vs]
values = [f"{tj_to_kwh_per_day_per_person(v.get('tj'))} kWh/day" for v in vs]
print(f"{list(zip(sectors,values))}")
```

**Q: List the UK's top 4-5 potential renewable energy sources in kWh/day per person**

For this we take the raw data from the [USwitch report here](https://www.uswitch.com/gas-electricity/studies/renewable-statistics/) from which we have the following figures in order for 2023:

```python
def gwh_per_year_to_kwh_per_day_per_person(gwh_per_year, population=68_000_000):
    """
    Convert energy from Gigawatt-hours per year (GWh/year) to Kilowatt-hours per day per person (KWh/day/person).
    
    Parameters:
    gwh_per_year (float): Energy in Gigawatt-hours per year.
    population (int): Population of the country. Default is 68 million.
    
    Returns:
    float: Energy in Kilowatt-hours per day per person, rounded to 2 decimal places.
    """
    # Convert GWh/year to KWh/year
    kwh_per_year = gwh_per_year * 10**6
    
    # Convert KWh/year to KWh/day
    kwh_per_day = kwh_per_year / 365
    
    # Convert to KWh per day per person
    kwh_per_day_per_person = kwh_per_day / population
    
    # Round the result to 2 decimal places
    return round(kwh_per_day_per_person, 2)
```

### Week 7: Biomass Energy

**Q: If air flows through a pipe at room temperature (20$^o$C) and 150000Pa, calculate the following in $mol/m^3$: a) concentration of air in the tube, b) concentration of oxygen in the tube.**

The ideal gas law describes how the amount of gas is independent on volume ($V$), pressure ($P$) and temperature ($T$).  It can be expressed as follows:

$PV = nRT$ 

where:<br>
$R$ = gas constant equal to 8.314 $J/mol K$<br>
$n$ = number of moles

```python
def getConcentration(P,T):
    R = 8.314 # J/molK
    # concentration is n/V so by rearranging terms:
    concentration = P/(R*T)
    return round(concentration,1)

def celsiusToKelvin(temp):
    kelvin = temp + 273.15
    return kelvin

getConcentration(1.5e5,celsiusToKelvin(20))
```

### Week 8: Electricity Network Integration

**Q: Three generators operate at a rated frequency of 60 Hz. Their rated powers are 300 MW, 500 MW and 600 MW, respectively. Let's assume that several large loads suddenly disconnected from the system, which caused an increase in the frequency of 0.3 Hz. Through the application of primary control, each generator ended up reducing their supply by -30 MW, -62.5 MW and -100 MW, respectively.
Based on this reduction, what is the droop characteristic of each of these generators?
(In percentage, without decimals, for example if the answer is 10%, then write it as 10)**

$\large R_{gi} = \large\frac{\frac{\Delta f}{f_r}\}{\frac{\Delta P_{gi}}{P_{gi,r}}}\$

```python
def droop(ratedPower,ratedFreq,deltaPower,deltaFrequency):
    Rgi = -(deltaFrequency/ratedFreq) / (deltaPower/ratedPower)
    return round(Rgi*100,0)

droop(300,60,-30,+0.3)
```

```python
droop(500,60,-62.5,+0.3)
```

```python
droop(600,60,-100,+0.3)
```

**Q: What is the network power frequency characteristic λ?**

```python
def powerFreqCharacteristic(deltaPowers,deltaFrequency): 
    powerChange = -sum(deltaPowers)
    return round(powerChange/deltaFrequency, 2)

powerFreqCharacteristic([-30,-62.5,-100], 0.3)
```

**Q: In the lecture it was mentioned that the energy density of hydrogen in a 700 bar tank is 5 MJ/kg. We want to use this type of tank for a hydrogen fuelled car. Calculate the weight of the tank (in kg) that is needed for such a car. Use the following information: i) the car has a driving range of 500 km, ii) the car needs 10 kWh/100km electrical energy, iii) the fuel cell in the car has an efficiency of 60%.**

Car needs to store 50kWh/0.6 of hydrogen. 1kWh =  3600000J

```python
def requiredHydrogenEnergy(rng,electricalEnergyPer100km,efficiency):
    fuelEnergy = (rng/100 * electricalEnergyPer100km)/efficiency
    return fuelEnergy

def convertkWhToMJ(kWh):
    ''' 1kWh = 3.6MJ '''
    return int(round(kWh * 3.6,0))

def weightOfTank(MJ, energyDensity):
    return int(round(MJ/energyDensity,0))

weightOfTank(convertkWhToMJ(requiredHydrogenEnergy(500, 10, 0.6)),5)
```

Answer is 60kg.
